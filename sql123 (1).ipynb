# query-1

```python
import pandas as pd
import sqlite3 as sql
database="imdb1.db"
con=sql.connect(database)
```

```python
query="""select a.name,b.title,b.year from (select a.mid,b.Name from (select pid,mid from m_director where mid in (select mid from m_genre where gid in (select gid from genre where Name Like '%Comedy%'))) a inner join person b on a.pid=b.pid) a inner join (select * from movie where (year%400=0 or (year%100!=0 and year%4=0 ))) b on b.mid=a.mid """
df1=pd.read_sql_query(query,con)
df1
```

# query-7

```python
query="""select a.year as start_year_decade, a.year + 9 as end_of_decade ,count(*) as num_movies from (select distinct year from movie) a join movie b
on b.year >= a.year and b.year < a.year + 10
group by a.year
order by count(*) desc
limit 1
"""
df7=pd.read_sql_query(query,con)
df7
```
